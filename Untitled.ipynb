```python
# Célula 1: Importações
import numpy as np
from scipy.ndimage import shift
from sklearn.datasets import fetch_openml
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

print("Bibliotecas importadas com sucesso!")
```

```python
# Célula 2: Carregar os Dados
mnist = fetch_openml('mnist_784', version=1, as_frame=False, parser='liac-arff')
X, y = mnist["data"], mnist["target"]
y = y.astype(np.uint8)

# Dividir entre treino e teste
# Usaremos os primeiros 60.000 para treinar e os 10.000 restantes para testar
X_train, X_test, y_train, y_test = X[:60000], X[60000:], y[:60000], y[60000:]

print("Dados carregados. Formato do conjunto de treino:", X_train.shape)
print("Formato do conjunto de teste:", X_test.shape)
```

```python
# Célula 3: Definir a Função para Mover a Imagem

from scipy.ndimage import shift # Importação específica para a função

# A função abaixo atende ao requisito de criar uma função para mover as imagens do MNIST. 
def mover_imagem(imagem, dx, dy):
  """
  Move uma imagem (vetor de 784 pixels) em dx (horizontal) e dy (vertical) pixels.
  """
  # Remodela o vetor para uma matriz 2D (28x28)
  imagem = imagem.reshape((28, 28))
  
  # Desloca a imagem usando a função shift, conforme sugerido. 
  # cval=0 garante que os espaços vazios sejam preenchidos com preto. 
  imagem_deslocada = shift(imagem, [dy, dx], cval=0, mode='constant')
  
  # Achata a matriz de volta para um vetor de 784 pixels
  return imagem_deslocada.reshape([-1])

print("Função 'mover_imagem' criada com sucesso!")
```

```python
# Célula 4: Aumentar o Conjunto de Treinamento

print("Iniciando o aumento de dados... Isso pode levar alguns minutos.")

imagens_aumentadas = []
rotulos_aumentados = []

# Itera sobre cada imagem e rótulo do conjunto de treino original
for imagem, rotulo in zip(X_train, y_train):
    # 1. Adiciona a imagem original
    imagens_aumentadas.append(imagem)
    rotulos_aumentados.append(rotulo)

    # 2. Cria e adiciona as 4 versões deslocadas, como pedido na avaliação 
    # Move para a direita
    imagens_aumentadas.append(mover_imagem(imagem, dx=1, dy=0))
    rotulos_aumentados.append(rotulo)

    # Move para a esquerda
    imagens_aumentadas.append(mover_imagem(imagem, dx=-1, dy=0))
    rotulos_aumentados.append(rotulo)

    # Move para baixo
    imagens_aumentadas.append(mover_imagem(imagem, dx=0, dy=1))
    rotulos_aumentados.append(rotulo)

    # Move para cima
    imagens_aumentadas.append(mover_imagem(imagem, dx=0, dy=-1))
    rotulos_aumentados.append(rotulo)

# Converte as listas para arrays numpy
X_train_aumentado = np.array(imagens_aumentadas)
y_train_aumentado = np.array(rotulos_aumentados)

# Embaralha o novo conjunto de dados para garantir que o modelo não aprenda em sequência
indices_embaralhados = np.random.permutation(len(X_train_aumentado))
X_train_aumentado = X_train_aumentado[indices_embaralhados]
y_train_aumentado = y_train_aumentado[indices_embaralhados]

print("Aumento de dados concluído!")
print("Tamanho do novo conjunto de treino:", X_train_aumentado.shape)
```

```python
# Célula 5 (VERSÃO OTIMIZADA)

print("Iniciando o treinamento do modelo SGD usando todos os núcleos da CPU...")

# n_jobs=-1 diz ao classificador para usar todos os processadores disponíveis
sgd_clf = SGDClassifier(random_state=42, n_jobs=-1) 
sgd_clf.fit(X_train_aumentado, y_train_aumentado)

print("Treinamento concluído!")
```

```python
# Célula 6: Avaliar o Modelo e a "Confusão entre os Números"

print("Fazendo predições no conjunto de teste...")
y_pred = sgd_clf.predict(X_test)

# Calcula a acurácia
acuracia = accuracy_score(y_test, y_pred)
print(f"\nAcurácia final do modelo com dados aumentados: {acuracia:.4f}")

# Gera a matriz de confusão para avaliar o desempenho.
print("Gerando a Matriz de Confusão...")
matriz_confusao = confusion_matrix(y_test, y_pred)

# Plota a matriz de confusão para uma melhor visualização.
# O objetivo aqui é verificar se a confusão entre os números diminuiu.
plt.figure(figsize=(10, 7))
sns.heatmap(matriz_confusao, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Valores Preditos')
plt.ylabel('Valores Reais')
plt.title('Matriz de Confusão do Modelo com Dados Aumentados')
plt.show()
```
